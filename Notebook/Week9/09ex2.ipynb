```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 演習2
>前の演習の表に基づき、東京の各気温の変化を棒グラフで表すプログラムを作成せよ

```python
X = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
X0 = np.arange(len(X)) 
Y1 = [9.6,10.4,13.6,19.0,22.9,25.5,29.2,30.8,26.9,21.5,16.3,11.9]
Y2 = [5.2,5.7,8.7,13.9,18.2,21.4,25.0,26.4,22.8,17.5,12.1,7.6]
Y3 = [0.9,1.7,4.4,9.4,14.0,18.0,21.8,23.0,19.7,14.2,8.3,3.5]

w = 0.4

plt.bar(X0-0.2,Y1,w,label='High')  
plt.bar(X0,Y2,w,label='Mean')
plt.bar(X0+0.2,Y3,w,label='Low')   

plt.xticks(X0,X)   
plt.title('Temperatures in Tokyo(1981-2010)')
plt.xlabel('Month')
plt.ylabel('Celsius degree')
plt.legend()
plt.show()
```
