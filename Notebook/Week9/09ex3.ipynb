````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

## 演習3
>ある商店の販売員4名の今月の売り上げ結果が以下の表に示されている。売り上げの販売員ごとの内訳も分かるような以下の販売実績グラフを表示するプログラムを作成せよ。ただし以下の変数を使うこと。
>```Python
>sales = [[100,410,120],[430,120,20],[190,184,420],[330,90,280]]
>types = ['Office','Home','Hobby']
>persons = ['Sato','Ueda','Mori','Suzuki']
>```

````python
sales = [[100,410,120],[430,120,20],[190,184,420],[330,90,280]]
types = ['Office','Home','Hobby']
persons = ['Sato','Ueda','Mori','Suzuki']

w = 0.5  
x = np.arange(len(types)) 

C = np.array(sales[3])
BC = np.array(sales[2])+ C
ABC = np.array(sales[1])+ BC


plt.bar(x, sales[0], width=w, label=persons[0],bottom=ABC)
plt.bar(x, sales[1], width=w, label=persons[1],bottom=BC)
plt.bar(x, sales[2], width=w, label=persons[2],bottom=C)
plt.bar(x, sales[3], width=w, label=persons[3])

plt.ylim(0,1500) 
plt.xticks(x,types)
plt.title('Sales Results')
plt.xlabel('Type of Products')
plt.ylabel('Amount of Sales')
plt.legend(loc = "upper right")
plt.show()
````
